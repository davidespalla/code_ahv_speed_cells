# Binomial tests

Test of independence of couple of features
N= total recorded cells
PA= cell in population 1
PB= cell in population 2
p=pA*pB cell in the intersection if independent
k= observed cells
pval= probability of observing k given p (under the independence assumption)

## Grid x AHV

```python
from scipy.stats import binom_test
N=len(cellMetadata[cellMetadata["include"]==1])
pA=len(cellMetadata[cellMetadata["AHVModulated"]==1])/float(N)
pB=len(cellMetadata[cellMetadata["gridModulated"]==1])/float(N)
p=pA*pB
k=len(cellMetadata[(cellMetadata["AHVModulated"]==1)&(cellMetadata["gridModulated"]==1)])
print("pval="+str(binom_test(k,N,p=p))+" expected: "+str(p*N)+" observed: "+str(k))
```

## HD x AHV

```python
from scipy.stats import binom_test
N=len(cellMetadata[cellMetadata["include"]==1])
pA=len(cellMetadata[cellMetadata["AHVModulated"]==1])/float(N)
pB=len(cellMetadata[cellMetadata["HDModulated"]==1])/float(N)
p=pA*pB
k=len(cellMetadata[(cellMetadata["AHVModulated"]==1)&(cellMetadata["HDModulated"]==1)])
print("pval="+str(binom_test(k,N,p=p))+" expected: "+str(p*N)+" observed: "+str(k))
```

## HD x Gird x AHV

```python
from scipy.stats import binom_test
N=len(cellMetadata[cellMetadata["include"]==1])
pA=len(cellMetadata[cellMetadata["AHVModulated"]==1])/float(N)
pB=len(cellMetadata[cellMetadata["HDModulated"]==1])/float(N)
pC=len(cellMetadata[cellMetadata["gridModulated"]==1])/float(N)
p=pA*pB*pC
k=len(cellMetadata[(cellMetadata["AHVModulated"]==1)&(cellMetadata["HDModulated"]==1)&(cellMetadata["gridModulated"]==1)])
print("pval="+str(binom_test(k,N,p=p))+" expected: "+str(p*N)+" observed: "+str(k))
```
