```python
import pandas as pd
#import seaborn as sns
from scipy.stats import pearsonr
from scipy import stats 
from scipy.ndimage import gaussian_filter
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from scipy.optimize import curve_fit
#sns.set_style("white")
```

# Tuning curve shape

```python
#DATA import
angularVelocity=pd.read_csv("TrialData/filteredAngularVelocity.csv")
speed=pd.read_csv("TrialData/filteredSpeed.csv")

trialMetadata=pd.read_csv("TrialData/metadata.csv",index_col=0)
cellMetadata=pd.read_csv("CellData/metadata.csv",index_col=0)

rate=pd.read_csv("CellData/firingRate.csv")

angularVelocity.columns=trialMetadata["trialIndex"].values
speed.columns=trialMetadata["trialIndex"].values
```

```python
def r2linear(x,y):
    bad = ~np.logical_or(np.isnan(x), np.isnan(y))
    a=np.compress(bad, x)  
    b=np.compress(bad, y)
    b=b/max(b)
    X=asmatrix(a).T
    regr = linear_model.LinearRegression()
    regr.fit(X,b)
    y_pred = regr.predict(X)
    #mse=mean_squared_error(y, y_pred)
    r2=r2_score(b,y_pred)
    return r2

def linearFit(x,y):
    bad = ~np.logical_or(np.isnan(x), np.isnan(y))
    a=np.compress(bad, x)  
    b=np.compress(bad, y)
    b=b
    X=asmatrix(a).T
    regr = linear_model.LinearRegression()
    regr.fit(X,b)
    y_pred = regr.predict(X)
    return y_pred

def fsigmoid(x, a, b):
    return 1.0 / (1.0 + np.exp(-a*(x-b)))


def r2sigmoid(x,y):
    bad = ~np.logical_or(np.isnan(x), np.isnan(y))
    a=np.compress(bad, x)  
    b=np.compress(bad, y)
    b=b/max(b)
    popt, pcov = curve_fit(fsigmoid,a,b, method='dogbox')
    y_pred=fsigmoid(a,popt[0],popt[1])
    r2=r2_score(b,y_pred)
    return r2

def sigmoidFit(x,y):
    bad = ~np.logical_or(np.isnan(x), np.isnan(y))
    a=np.compress(bad, x)  
    b=np.compress(bad, y)
    b=b/max(b)
    popt, pcov = curve_fit(fsigmoid, a,b, method='dogbox')
    y_pred=fsigmoid(a,popt[0],popt[1])
    return y_pred
```

# SPEED

```python
# FITTING
speedCells=cellMetadata[(cellMetadata["include"]==1)&(cellMetadata["speedModulated"]==1)]
lin_r2=[]
sig_r2=[]
for cell in range(len(cellMetadata.index)):
    if cellMetadata["include"].values[cell]==1 and cellMetadata["speedModulated"].values[cell]==1:
        r=rate[cellMetadata.index[cell]].values
        s=speed[cellMetadata["trialIndex"][cell]].values
        bad = ~np.logical_or(np.isnan(r), np.isnan(s))
        a=np.compress(bad, s)  
        b=np.compress(bad, r)
        x=linspace(2,50,20)
        y=stats.binned_statistic(a,b, statistic='mean', bins=20, range=[2,50])[0]
        lin_r2.append(r2linear(x,y))
        try:
            sig_r2.append(r2sigmoid(x,y))
        except:
            sig_r2.append(np.nan)
    else:
        lin_r2.append(np.nan)
        sig_r2.append(np.nan)
```

```python
# Append R2 in metadata
cellMetadata["speedLinearR2"]=lin_r2
cellMetadata["speedSigmoidR2"]=sig_r2
```

## Plotting: Linear

```python
region="MEC"
speedScoreSorted=cellMetadata[(cellMetadata["speedSigmoidR2"]<cellMetadata["speedLinearR2"])&(cellMetadata["region"]==region)].sort_values("speedLinearR2",ascending=False)

for cell in range(5):
    #add line below to divide by region
    #speedScoreSorted=cellMetadata[cellMetadata["region"]=="PrS"].sort_values("speedScore",ascending=False)
    r=rate[speedScoreSorted.index[cell]].values
    s=speed[speedScoreSorted["trialIndex"][cell]].values
    figure(figsize=(6,6))
    bad = ~np.logical_or(np.isnan(r), np.isnan(s))
    a=np.compress(bad, s)  
    b=np.compress(bad, r)
    b=b
    x=linspace(2,50,20)
    y=stats.binned_statistic(a,b, statistic='mean', bins=20, range=[2,50])[0]
    counts=stats.binned_statistic(a,b, statistic='count', bins=20, range=[2,50])[0]
    std=stats.binned_statistic(a,b, statistic='std', bins=20, range=[2,50])[0]
    error=std/sqrt(counts)
    yp=linearFit(x,y)
    #y1=gaussian_filter(y,1)
    plot(x,y,linewidth=3)
    plt.fill_between(x, y-error, y+error,alpha=0.5)
    plot(x,yp,linewidth=3)
    xlabel("speed (cm/s) \n score= "+str(speedScoreSorted["speedScore"][cell]))
    ylabel("Firing rate (Hz)")
    xlim((0,52))
    savefig(region+"speed_tuning_linear_cell"+str(cell+1)+".svg")
    
```

## Plotting:Sigmoid

```python
region="MEC"
speedScoreSorted=cellMetadata[(cellMetadata["speedSigmoidR2"]>cellMetadata["speedLinearR2"])&(cellMetadata["region"]==region)].sort_values("speedSigmoidR2",ascending=False)

for cell in range(5):
    #add line below to divide by region
    #speedScoreSorted=cellMetadata[cellMetadata["region"]=="PrS"].sort_values("speedScore",ascending=False)
    r=rate[speedScoreSorted.index[cell]].values
    s=speed[speedScoreSorted["trialIndex"][cell]].values
    figure(figsize=(6,6))
    bad = ~np.logical_or(np.isnan(r), np.isnan(s))
    a=np.compress(bad, s)  
    b=np.compress(bad, r)
    x=linspace(2,50,20)
    y=stats.binned_statistic(a,b, statistic='mean', bins=20, range=[2,50])[0]
    counts=stats.binned_statistic(a,b, statistic='count', bins=20, range=[2,50])[0]
    std=stats.binned_statistic(a,b, statistic='std', bins=20, range=[2,50])[0]
    error=std/sqrt(counts)
    yp=max(y)*sigmoidFit(x,y)
    #error/max(y)
    #y=y/max(y)
    #y1=gaussian_filter(y,1)
    plot(x,y,linewidth=3)
    plot(x,yp,linewidth=3)
    plt.fill_between(x, y-error, y+error,alpha=0.5)
    xlabel("speed (cm/s) \n score= "+str(speedScoreSorted["speedScore"][cell]))
    ylabel("Firing rate (Hz)")
    xlim((0,52))
    savefig(region+"speed_tuning_sigmoid_cell"+str(cell+1)+".svg")
```

# AHV ABS

The fit is performed on the average of the right part of the plot and the mirrored left part of the plot

```python
absAngularVelocity=abs(angularVelocity)
```

```python
# FITTING
angularVelocityCells=cellMetadata[(cellMetadata["include"]==1)&(cellMetadata["absAngularVelocityModulated"]==1)]
lin_r2=[]
sig_r2=[]
for cell in range(len(cellMetadata.index)):
    if cellMetadata["include"].values[cell]==1 and cellMetadata["absAngularVelocityModulated"].values[cell]==1:
        r=rate[cellMetadata.index[cell]].values
        s=angularVelocity[cellMetadata["trialIndex"][cell]].values
        bad = ~np.logical_or(np.isnan(r), np.isnan(s))
        a=np.compress(bad, s)  
        b=np.compress(bad, r)
        x=linspace(0,3,10)
        yr=stats.binned_statistic(a,b, statistic='mean', bins=10, range=[0,3])[0]
        yl=stats.binned_statistic(a,b, statistic='mean', bins=10, range=[-3,0])[0]
        y=(yr+yl[::-1])/2.0
        lin_r2.append(r2linear(x,y))
        try:
            sig_r2.append(r2sigmoid(x,y))
        except:
            sig_r2.append(np.nan)
    else:
        lin_r2.append(np.nan)
        sig_r2.append(np.nan)
```

```python
# Append R2 in metadata
cellMetadata["angularVelocityABSLinearR2"]=lin_r2
cellMetadata["angularVelocityABSigmoidR2"]=sig_r2
```

## Plotting: Linear

```python
region="MEC"
angularVelocityScoreSorted=cellMetadata[(cellMetadata["angularVelocityABSigmoidR2"]<cellMetadata["angularVelocityABSLinearR2"])&(cellMetadata["region"]==region)].sort_values("angularVelocityABSLinearR2",ascending=False)
for cell in range(5):
    #add line below to divide by region
    #angularVelocityScoreSorted=cellMetadata[cellMetadata["region"]=="PrS"].sort_values("angularVelocityScore",ascending=False)
    r=rate[angularVelocityScoreSorted.index[cell]].values
    s=angularVelocity[angularVelocityScoreSorted["trialIndex"][cell]].values
    figure(figsize=(6,6))
    bad = ~np.logical_or(np.isnan(r), np.isnan(s))
    a=np.compress(bad, s)  
    b=np.compress(bad, r)
    b=b
    x=linspace(0,3,10)
    yr=stats.binned_statistic(a,b, statistic='mean', bins=10, range=[0,3])[0]
    yl=stats.binned_statistic(a,b, statistic='mean', bins=10, range=[-3,0])[0]
    y=(yr+yl[::-1])/2
    countsr=stats.binned_statistic(a,b, statistic='count', bins=10, range=[0,3])[0]
    countsl=stats.binned_statistic(a,b, statistic='count', bins=10, range=[-3,0])[0]
    stdr=stats.binned_statistic(a,b, statistic='std', bins=10, range=[0,3])[0]
    stdl=stats.binned_statistic(a,b, statistic='std', bins=10, range=[-3,0])[0]
    errorR=stdr/sqrt(countsr)
    errorL=stdl/sqrt(countsl)
    error=sqrt((pow(errorR,2)+pow(errorL[::-1],2))/2.0) # TO CHECK
    yp=linearFit(x,y)
    
    plot(x,y,linewidth=3)
    plt.fill_between(x, y-error, y+error,alpha=0.5)
    plot(x,yp,linewidth=3)
    xlabel(" angularVelocity (cm/s) \n score= "+str(angularVelocityScoreSorted["absAngularVelocityScore"][cell]))
    ylabel("Firing rate (Hz)")
    savefig(region+"angularVelocityABS_Linear_cell"+str(cell+1)+".svg")
```
