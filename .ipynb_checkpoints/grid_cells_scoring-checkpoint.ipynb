```python
%pylab inline
import scipy.ndimage.interpolation as rot
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr
from scipy import stats 
from scipy.ndimage import gaussian_filter
sns.set_style("white")
```

```python
def calculateRatemap(x,y,rate,dim):
    binwidth=2 #2 cm bins
    xrange=dim
    yrange=dim
    nbinsx=int(xrange/binwidth)
    nbinsy=int(yrange/binwidth)
    sigma=2 #2 bins = 4 cms
    bad = ~np.logical_or(np.isnan(x), np.isnan(y))
    a=np.compress(bad, x)  
    b=np.compress(bad, y)
    c=compress(bad,rate)
    rawmap=stats.binned_statistic_2d(a,b,c, statistic='mean', bins=[nbinsx,nbinsy],range=None)[0]
    rawmap[np.isnan(rawmap)]=0
    smoothedmap=gaussian_filter(rawmap,sigma)
    return smoothedmap

def autocorrelation(rate_map, mode='full', normalize=True):
    return fftcorrelate2d(rate_map, rate_map, mode=mode, normalize=normalize)

def fftcorrelate2d(arr1, arr2, mode='full', normalize=True):
    from scipy.signal import fftconvolve
    if normalize:
        a_ = np.reshape(arr1, (1, arr1.size))
        v_ = np.reshape(arr2, (1, arr2.size))
        arr1 = (arr1 - np.mean(a_)) / (np.std(a_) * len(a_))
        arr2 = (arr2 - np.mean(v_)) / np.std(v_)
    corr = fftconvolve(arr1, np.fliplr(np.flipud(arr2)), mode=mode)
    return corr



def mat_sqdist(n): #n,n is the center, autoc is 2n+1,2n+1, return mat with squared distances from center
    mat_dist = np.zeros((2*n+1,2*n+1))
    for i in range(n+1):
        for j in range(n+1):
            a = (n-i)**2 + (n-j)**2
            mat_dist[i,j]=mat_dist[2*n-i,j]=mat_dist[i,2*n-j]=mat_dist[2*n-i,2*n-j]= a
    return mat_dist

def gridscore(ac,mat_sds, lags = np.array([-3,0,3])): #faster! not so many rotations
    def corr_raddeg(ac,acr,ind_rad):
         return np.corrcoef(ac[ind_rad],acr[ind_rad])[0,1]
    def gridscore_rad(ac,ac_rots,ind_rad):
        g60 = max([corr_raddeg(ac,ac_rot,ind_rad) for ac_rot in ac_rots[0]])
        g120 = max([corr_raddeg(ac,ac_rot,ind_rad) for ac_rot in ac_rots[1]])
        g30 = min([corr_raddeg(ac,ac_rot,ind_rad) for ac_rot in ac_rots[2]])
        g90 = min([corr_raddeg(ac,ac_rot,ind_rad) for ac_rot in ac_rots[3]])
        g150 = min([corr_raddeg(ac,ac_rot,ind_rad) for ac_rot in ac_rots[4]])
        return 1/2*(g60+g120) - 1/3*(g30 + g90 + g150)
    
    
    n = (ac.shape[0] - 1)/2 #n,n is the center
    max_grs = -100          #max grid score initialized at arbitrary very low value
    
    rots=[]
    
    rots.append([rot.rotate(ac,deg,reshape=False) for deg in 60 + lags])#0
    rots.append([rot.rotate(ac,deg,reshape=False) for deg in 120 + lags])#1
    rots.append([rot.rotate(ac,deg,reshape=False) for deg in 30 + lags])#2
    rots.append([rot.rotate(ac,deg,reshape=False) for deg in 90 + lags])#3
    rots.append([rot.rotate(ac,deg,reshape=False) for deg in 150 + lags])#4
    if np.mean(ac) != 0:
        for rad in range(int(2*n/5),int((n/5)*4.5),2):
            ind_rad = (mat_sds<=(rad**2)) * (mat_sds>=((n/5)**2)) * (ac != 0)
            max_grs = max(max_grs,gridscore_rad(ac,rots,ind_rad))
    else:
        max_grs=0
    return max_grs

def calculatePlotRatemap(x,y,rate,dim):
    binwidth=2 #2 cm bins
    xrange=dim
    yrange=dim
    nbinsx=int(xrange/binwidth)
    nbinsy=int(yrange/binwidth)
    sigma=2 #2 bins = 4 cms
    bad = ~np.logical_or(np.isnan(x), np.isnan(y))
    a=np.compress(bad, x)  
    b=np.compress(bad, y)
    c=compress(bad,rate)
    rawmap=stats.binned_statistic_2d(a,b,c, statistic='mean', bins=[nbinsx,nbinsy],range=None)[0]
    rawmap[np.isnan(rawmap)]=0
    smoothedmap=gaussian_filter(rawmap,sigma)
    return smoothedmap
```

## GRID SCORING

```python
cellMetadata=pd.read_csv("CellData/metadata.csv",index_col=0)
trialMetadata=pd.read_csv("TrialData/metadata.csv",index_col=0)
xs=pd.read_csv("TrialData/x1.csv")
ys=pd.read_csv("TrialData/y1.csv")
rate=pd.read_csv("CellData/firingRate.csv")
xs.columns=trialMetadata["trialIndex"].values
ys.columns=trialMetadata["trialIndex"].values
```

```python
gridScore=np.zeros(len(rate.T))
for i in range(len(rate.columns)):
        trialId=cellMetadata.loc[rate.columns[i],:]["trialIndex"]
        if cellMetadata["include"][rate.columns[i]]==1:
            x= xs[trialId].values
            y = ys[trialId].values
            r = rate[rate.columns[i]].values
            dim=trialMetadata[trialMetadata["trialIndex"]==trialId]["dim"]
            bindim=int(dim/2) #if the bins are 2 cms wide
            rate_map=calculateRatemap(x,y,r,dim)
            ac=autocorrelation(rate_map, mode='full', normalize=True)
            mat_sqs = mat_sqdist(bindim-1)
            gridScore[i]=gridscore(ac,mat_sqs)
        else:
            gridScore[i]=np.nan
        if i %100==0:
            print("Scored cell: "+str(i))
cellMetadata["gridScore"]=gridScore
```

```python
#export
cellMetadata.to_csv("CellData/metadata.csv")
```

```python
figure(figsize=(8,8))
hist([x for x in gridScore if not isnan(x)],bins=30);
xlabel("Grid Score",fontsize=20)
ylabel("# of cells",fontsize=20)
#savefig("imgs/speedScore_distr.png")
```

# Conjunctive Plots

```python
%pylab inline
import scipy.ndimage.interpolation as rot
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
from scipy import stats 
from scipy.ndimage import gaussian_filter
from matplotlib.colors import LinearSegmentedColormap
```

```python
xs=pd.read_csv("TrialData/x1.csv")
ys=pd.read_csv("TrialData/y1.csv")
angularVelocity=pd.read_csv("TrialData/filteredAngularVelocity.csv")
speed=pd.read_csv("TrialData/filteredSpeed.csv")

trialMetadata=pd.read_csv("TrialData/metadata.csv",index_col=0)
cellMetadata=pd.read_csv("CellData/metadata.csv",index_col=0)

rate=pd.read_csv("CellData/firingRate.csv")

angularVelocity.columns=trialMetadata["trialIndex"].values
xs.columns=trialMetadata["trialIndex"].values
ys.columns=trialMetadata["trialIndex"].values
speed.columns=trialMetadata["trialIndex"].values
```

```python
gridScoreSorted=cellMetadata[(cellMetadata["region"]=="PrS")&(cellMetadata["angularVelocityCCWModulated"]==1)].sort_values("gridScore",ascending=False)

interval = np.hstack([np.linspace(0, 0.4), np.linspace(0.5, 1)])
colors = plt.cm.jet(interval)
cmap = LinearSegmentedColormap.from_list('name', colors)

for cell in range(5):
    r=rate[gridScoreSorted.index[cell]].values
    x=xs[gridScoreSorted["trialIndex"][cell]].values
    y=ys[gridScoreSorted["trialIndex"][cell]].values
    s=angularVelocity[gridScoreSorted["trialIndex"][cell]].values
    dim=trialMetadata[trialMetadata["trialIndex"]==gridScoreSorted["trialIndex"][cell]]["dim"].values
    rate_map=calculatePlotRatemap(x,y,r,dim)
    ac=autocorrelation(rate_map, mode='full', normalize=True)/len(rate_map.flatten())
    figure(figsize(15,5))
    ax=subplot(1,3,1)
    rim=ax.imshow(rate_map,cmap=cmap)
    #colorbar(rim)
    xlabel("peak rate="+str(max(rate_map.flatten())))
    ax=subplot(1,3,2)
    acim=ax.imshow(ac,cmap=cm.jet)
    xlabel("grid score="+str(gridScoreSorted["gridScore"][cell]))
    #colorbar(acim)
    ax=subplot(1,3,3)
    bad = ~np.logical_or(np.isnan(r), np.isnan(s))
    a=np.compress(bad, s)  
    b=np.compress(bad, r)
    x=linspace(-3,3,10)
    y=stats.binned_statistic(a,b, statistic='mean', bins=10, range=[-3,3])[0]
    y1=gaussian_filter(y,2)
    plot(x,y1,linewidth=3)
    xlabel("angular velocity (rad/s) \n score="+str(gridScoreSorted["angularVelocityScore"][cell]))
    ylabel("Firing rate (Hz)")
    tight_layout()
    savefig("PRS_Grid_x_CCWangularVelocity_cell"+str(cell+1)+".svg")
```
