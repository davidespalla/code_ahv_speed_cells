```python
%pylab inline
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr
sns.set()
```

# Speed cells scoring

```python
#trial data
speed=pd.read_csv("TrialData/speed.csv")
speedMask=pd.read_csv("TrialData/speedFilterMask.csv")
trialMetadata=pd.read_csv("TrialData/metadata.csv",index_col=0)
#cells data
rate=pd.read_csv("CellData/firingRate.csv")
cellMetadata=pd.read_csv("CellData/metadata.csv",index_col=0)
speed.columns=trialMetadata["trialIndex"].values
speedMask.columns=trialMetadata["trialIndex"].values
```

```python
def speedFilter(array,mask):
    out=array
    for i in range(len(array)):
        if mask[i]==0:
            out[i]=np.nan
    return out
```

```python
#select cells and trials for analysis
speedScore=np.zeros(len(rate.T))
for i in range(len(rate.columns)):
        trialId=cellMetadata.loc[rate.columns[i],:]["trialIndex"]
        if cellMetadata["include"][rate.columns[i]]==1:
            x = rate[rate.columns[i]].values
            y = speedFilter(speed[trialId].values,speedMask[trialId])
            bad = ~np.logical_or(np.isnan(x), np.isnan(y))
            a=np.compress(bad, x)  
            b=np.compress(bad, y) 
            speedScore[i]=pearsonr(a,b)[0]
        else:
            speedScore[i]=np.nan
        if i %100==0:
            print("Scored cell: "+str(i))
cellMetadata["speedScore"] = speedScore
```

```python
#export
cellMetadata.to_csv("CellData/metadata.csv")
```

```python
figure(figsize=(8,8))
hist([x for x in speedScore if not isnan(x)]);
xlabel("Speed Score",fontsize=20)
ylabel("# of cells",fontsize=20)
savefig("imgs/speedScore_distr.png")
```
